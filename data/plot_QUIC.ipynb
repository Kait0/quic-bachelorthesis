```python
import datetime
print(datetime.datetime.now())
i = 0
list2 = []
for index, row in tcp13.iterrows(): #change tcp to tcp13 to compare quic with TLS 1.3
    url = quic['Url'] == row['Url']
    timelow = quic['TimeOfMeasurement'] - 60 <= row['TimeOfMeasurement']
    timehigh = row['TimeOfMeasurement'] <= quic['TimeOfMeasurement'] + 60
    t = quic[url & timelow & timehigh]
    t = t.reset_index(drop=True)
    if(t.empty != True):
        list2.append([t.at[0,'Url'], t.at[0,'Ip'], t.at[0,'Port'], (row['ConnectionEstablishmentTime'] - t.at[0,'ConnectionEstablishmentTime'])])
    i = i + 1
    if(i%1000 == 0): #Show me how fast the script is progressing
        print(datetime.datetime.now())
        
dfQuicTls12 = pd.DataFrame(list2, columns = ['URL','IP', 'PORT', 'ConnDiff'])
print(dfQuicTls12)
```

```python
x = np.sort(dfQuicTls12['ConnDiff'])
y = np.arange(1, len(x)+1)/float(len(x))

yticks = np.arange(0,1.1,0.2)
xticks = np.arange(-40,80,10)
fig = plt.figure(figsize=(5,1.5))
plt.rcParams['font.size'] = 12
plt.title('TLS1.3faster<--------------------->QUIC faster', fontsize=12, y=1.05)
ax = fig.axes[0]
ax.set_yticks(yticks)
ax.set_ylim([0,1.05])
ax.set_xlim([-40,70])
ax.set_xticks(xticks)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
_ = plt.plot(x,y, marker='s', linewidth=0.3, markersize=5, fillstyle='none', color = 'orange')
_ = plt.xlabel('latency difference in ipv4(x-axis is cut)')
_ = plt.ylabel('ECDF')
plt.axvline(x=0)

plt.legend(['Difference between Quic and TLS 1.3'], fontsize=7, loc='lower right')
plt.savefig('Plots\\emilia_v4_QUIC_3websites_differences.png', format='png', bbox_inches='tight')
plt.show()
```
